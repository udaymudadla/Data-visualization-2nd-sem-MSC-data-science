QUESTION 8: Which months in the year 2011 have the highest and the lowest total number of bikes rented? Repeat for the year 2012.

```python
month_names = {
    1:'january',
    2:'february',
    3:'March',
    4:'april',
    5:'may',
    6:'junr',
    7:'july',
    8:'august',
    9:'september',
    10:'october',
    11:'november',
    12:'december'
}
data_2011 = df[df['year'] == 2011]
year_2011 = data_2011.groupby('month')['count'].sum()

print('highest number of bike rented in 2011 =',
      month_names[year_2011.idxmax()], year_2011.max())
print('lowest number of bike rented in 2011 =',
      month_names[year_2011.idxmin()], year_2011.min())

data_2012 = df[df['year'] == 2012]
year_2012 = data_2012.groupby('month')['count'].sum()

print('highest number of bike rented in 2012 =',
      month_names[year_2012.idxmax()], year_2012.max())
print('lowest number of bike rented in 2012 =',
      month_names[year_2012.idxmin()], year_2012.min())
```

QUESTION 11: Create a new categorical column called day_period, which can take four possible values: night, morning, afternoon and evening. These values
correspond to the following binning of the hour column: 0-6: night, 6-12:
morning, 12-6: afternoon, 6-24:evening.

```python
def categorize_hour(hour):
    if 0 <= hour < 6:
        return 'night'
    elif 6 <= hour < 12:
        return 'morning'
    elif 12 <= hour < 18:    
        return 'afternoon'
    else:  # 18-24
        return 'evening'

df['day_period'] = df['hour'].apply(categorize_hour)
print(df['day_period'].value_counts().sort_index())
```

**QUESTION 12:** Generate a pivot table for the mean of the hourly total rentals count, with the 
index set to the day period and the column set to the working day column. 
What can you observe from the table ?

```python
pivot_table = df.pivot_table(values='count', 
                             index='day_period', 
                             columns='workingday',
                             aggfunc='mean')

pivot_table.columns = ['Non-working day', 'Working day']
period_order = ['night', 'morning', 'afternoon', 'evening']
pivot_table = pivot_table.reindex(period_order)

print(f"\nPivot table - Mean hourly rentals:")
print(pivot_table)
```

**Observations:**
1. Morning rentals are higher on working days (234.35) vs non-working days (157.77241379310345)
2. Afternoon rentals between working and non-working days, non working days are higher than the working days 
3. Evening rentals are higher on working days
4. Night rentals are consistently low regardless of working day status
